```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading all sheets won't help as those have different formats. 
# sheets = pd.read_excel("./Portfolio-Theory-Investment-Data-Monthly.xlsx",sheetname=["SPY","IWM","VWO","EFA","LQD","HYG","US Treasury Yields"],)
# Read each data set individually
#spy = pd.read_excel("./Portfolio-Theory-Investment-Data-Monthly.xlsx",sheetname=["SPY"],header=3)
sheets = pd.read_excel("./Portfolio-Theory-Investment-Data-Monthly.xlsx",sheetname=["SPY","IWM","VWO","EFA","LQD","HYG"],header=3)
spyds= sheets['SPY']
spyds= spyds.iloc[:102,:9]
spyds.head()
```

```python
iwmds=sheets['IWM']
iwmds=iwmds.iloc[:,:9]
iwmds.head()
```

```python
#VWO","EFA","LQD","HYG"
vwods=sheets['VWO']
vwods=vwods.iloc[:,:9]
vwods.head()
```

```python
#VWO","EFA","LQD","HYG"
efads=sheets['EFA']
efads=efads.iloc[:,:9]
efads.head()
```

```python
#VWO","EFA","LQD","HYG"
lqdds=sheets['LQD']
lqdds=lqdds.iloc[:,:9]
lqdds.head()
```

```python
#VWO","EFA","LQD","HYG"
hygds=sheets['HYG']
hygds=hygds.iloc[:,:9]
hygds.head()
```

```python
hygds.info()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20, 60))
spyds.plot(ax=axes[0,0],x='Date',y='Monthly Return',legend=True,title="SPY",color="red")
spyds.plot(ax=axes[0,0],x='Date',y='Excess Return',legend=True,title="SPY",color="blue")
iwmds.plot(ax=axes[1,0],x='Date',y='Monthly Return',legend=True,title="IWM",color="green")
iwmds.plot(ax=axes[1,0],x='Date',y='Excess Return',legend=True,title="IWM",color="red")
vwods.plot(ax=axes[2,0],x='Date',y='Monthly Return',legend=True,title="VWO",color="blue")
vwods.plot(ax=axes[2,0],x='Date',y='Excess Return',legend=True,title="VWO",color="yellow")
efads.plot(ax=axes[0,1],x='Date',y='Monthly Return',legend=True,title="EFA",color="yellow")
efads.plot(ax=axes[0,1],x='Date',y='Excess Return',legend=True,title="EFA",color="green")
lqdds.plot(ax=axes[1,1],x='Date',y='Monthly Return',legend=True,title="LQD",color="violet")
lqdds.plot(ax=axes[1,1],x='Date',y='Excess Return',legend=True,title="LQD",color="green")
hygds.plot(ax=axes[2,1],x='Date',y='Monthly Return',legend=True,title="HYG",color="orange")
hygds.plot(ax=axes[2,1],x='Date',y='Excess Return',legend=True,title="HYG",color="blue")
plt.show()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 25))
spyds.plot(ax=axes[0,0],x='Date',y='Adj Close',legend=True,title="SPY",color="red")
spyds.plot(ax=axes[0,0],x='Date',y='Close',legend=True,title="SPY",color="Black")
iwmds.plot(ax=axes[1,0],x='Date',y='Adj Close',legend=True,title="IWM",color="green")
iwmds.plot(ax=axes[1,0],x='Date',y='Close',legend=True,title="IWM",color="red")
vwods.plot(ax=axes[2,0],x='Date',y='Adj Close',legend=True,title="VWO",color="blue")
vwods.plot(ax=axes[2,0],x='Date',y='Close',legend=True,title="VWO",color="green")
efads.plot(ax=axes[0,1],x='Date',y='Adj Close',legend=True,title="EFA",color="yellow")
efads.plot(ax=axes[0,1],x='Date',y='Close',legend=True,title="EFA",color="red")
lqdds.plot(ax=axes[1,1],x='Date',y='Adj Close',legend=True,title="LQD",color="violet")
lqdds.plot(ax=axes[1,1],x='Date',y='Close',legend=True,title="LQD",color="yellow")
hygds.plot(ax=axes[2,1],x='Date',y='Adj Close',legend=True,title="HYG",color="orange")
hygds.plot(ax=axes[2,1],x='Date',y='Close',legend=True,title="HYG",color="blue")
plt.show()
```
