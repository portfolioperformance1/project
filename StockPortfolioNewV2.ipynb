```python
symbolsL=['spy','iwm','vwo','efa','lqd','hyg']
# Load the individual dataframes (such as spyds) into a dictionary so that they
# can be called by their symbol, ex. symbolsD['spy']
symbolsD={'spy':spyds, 'iwm':iwmds, 'vwo':vwods, 'efa':efads, 'lqd':lqdds, 'hyg':hygds}
symbolsD['spy'].head(3)
```

```python
# Make every ds include the same dates, even if some rows are filled with null

# temp1 is a dataframe with the same dates as spyds
temp1=pd.DataFrame(index=spyds.index.values)
temp1['Date']=spyds.Date.sort_values()
for sym in symbolsD.keys():
    symbolsD[sym] = pd.merge(temp1,symbolsD[sym], on='Date', how='left') #left join includes all rows from df on left    
    symbolsD[sym]['Year'] = symbolsD[sym]['Date'].apply(lambda x: int(str(x).split('-')[0]))
    symbolsD[sym]['Month'] = symbolsD[sym]['Date'].apply(lambda x: int(str(x).split('-')[1]))
vwods.head(13)
# vwods.tail(13)
# vwods[vwods['Year']=='2005']
symbolsD['vwo'].loc[symbolsD['vwo']['Year']==2005]
# Creating aliases so that we can call dataframes by their individual names (i.e. spyds) 
# and as elements of symbolsD['spy'] 
spyds=symbolsD['spy'] 
iwmds=symbolsD['iwm'] 
efads=symbolsD['efa']
lqdds=symbolsD['lqd'] 
hygds=symbolsD['hyg'] 
vwods=symbolsD['vwo']   #vwods is the same object as symbolsD['vwo]
vwods.head(3)
```
